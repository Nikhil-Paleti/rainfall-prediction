```python
import pandas as pd
```

```python
import imdlib as imd
```

```python
start_yr = 1951
end_yr = 2021
```

```python
data1 = imd.get_data('rain', start_yr, end_yr, fn_format='yearwise')
IPython.display.clear_output()
```

```python
data2 = imd.get_data('tmin', start_yr, end_yr, fn_format='yearwise')
IPython.display.clear_output()
```

```python
data3 = imd.get_data('tmax', start_yr, end_yr, fn_format='yearwise')
IPython.display.clear_output()
```

```python
import numpy as np
import os
import array
```

```python
# 1. Guwahati
# 26.1158° N, 91.7086° E

lat = 26.1158
lon = 91.7086
data1.to_csv('guwahati_rain.csv', lat, lon, "/content")
data2.to_csv('guwahati_tmin.csv', lat, lon, "/content")
data3.to_csv('guwahati_tmax.csv', lat, lon, "/content")

guwahati_rain = pd.read_csv("/content/guwahati_rain_26.12_91.71.csv")
guwahati_tmin = pd.read_csv("/content/guwahati_tmin_26.12_91.71.csv")
guwahati_tmax = pd.read_csv("/content/guwahati_tmax_26.12_91.71.csv")

guwahati = pd.DataFrame(columns=["Date", "Rain", "tMin", "tMax"])
guwahati["Date"] = guwahati_rain[guwahati_rain.columns[0]]
guwahati["Rain"] = guwahati_rain[guwahati_rain.columns[1]]
guwahati["tMin"] = guwahati_tmin[guwahati_tmin.columns[1]]
guwahati["tMax"] = guwahati_tmax[guwahati_tmax.columns[1]]

guwahati.to_csv("guwahati.csv", index=False)
guwahati.head()
```
